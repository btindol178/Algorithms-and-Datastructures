In a single toss of  fair (evenly-weighted) six-sided dice, find the probability that their sum will be at most 9.

```python
import pandas as pd
import numpy as np
```

```python
die1 = [1,2,3,4,5,6] 
die2 = [1,2,3,4,5,6]

df2 = pd.DataFrame(columns = ['x1','x2']) # make a dataframe with column for each die
for i in range(0,len(die1)):
    for j in range(0,len(die2)):
        var1 = int(die1[i]) # loop through each row in die 1 this will be 1 for 6 values of j
        var2 = int(die2[j]) # while in each value of i loop through all values of j to get each combination of i,j values
        df2 = df2.append({'x1':var1,'x2':var2}, ignore_index=True)

        
```

```python
die1 = [1,2,3,4,5,6] 
die2 = [1,2,3,4,5,6]

# probability that the sum greater than a certain number with two six sided die
def probability_of_roll_gt(value_gt):
    df2 = pd.DataFrame(columns = ['x1','x2']) # make a dataframe with column for each die
    for i in range(0,len(die1)):
        for j in range(0,len(die2)):
            var1 = int(die1[i]) # loop through each row in die 1 this will be 1 for 6 values of j
            var2 = int(die2[j]) # while in each value of i loop through all values of j to get each combination of i,j values
            df2 = df2.append({'x1':var1,'x2':var2}, ignore_index=True)
    df2['sum'] = df2['x1'] + df2['x2']
    a = df2.loc[df2['sum'] > value_gt,'sum'].count()/df2['sum'].count()
    a = round(a,3)
    return a
```

```python
probability_of_roll_gt(10)
```
