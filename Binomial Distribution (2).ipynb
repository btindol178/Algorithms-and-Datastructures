# BINOMIAL DISTRIBUTION

The ratio of boys to girls is 1.09:1 for babies born in Russia is .
    If there is 1 child born per birth, what proportion of Russian families with exactly 3 children will have at least  boys?



# Solution 1

```python
import math

# Why 2.09
# The ratio of boys:girls = 1.09/1 this means that if there are 1.09x boys,
# there are 1x girls (where the multiplying factor is x) the total number of boys
# and girls = 1.09x + 1x = 2.09 x the probability of getting a boy = 1.09x/2.09x = 1.09/2.09


# WHY  range 3,7 
#We need to find the proportion of at least 3 boys among 6, here, range 3-7 loop for 4 times with values 3, 4, 5, 6
def bi_dist(x, n, p):
    b = (math.factorial(n)/(math.factorial(x)*math.factorial(n-x)))*(p**x)*((1-p)**(n-x))
    return(b)


b, p, n = 0, 1.09/2.09, 6
for i in range(3,7):
    b += bi_dist(i, n, p)   
print("%.3f" %b)
```

# Solution 2

```python
import sys 
import math

b = [float(i) for i in sys.stdin.readline().split()]

def nCr(n,r):
    f = math.factorial
    return f(n)/( f(r) * f(n-r))

pb = b[0]/(b[0]+b[1])
pg = 1-pb

n = 6
p3b = 0 

for i in range(3,n+1):
    p3b += nCr(n,i) * (pb**i) * (pg**(n-i))

print(round(p3b,3))
```

# Solution 3

```python
import math


def binomial_dist(n,p,x):
    return (math.factorial(n) / (math.factorial(x)*math.factorial(n-x))) * math.pow(p,x) * math.pow(1-p,n-x)

# p_input, q_input = list(map(int, input().split()))
# p = p_input/(p_input+q_input)
p = 1.09/2.09
n = 6

probability_sum = 0
for x in range(3,7):
    probability_sum = probability_sum + binomial_dist(n=n, p=p, x=x)

print(round(probability_sum, 3))
```
