Task <br>
The probability that a machine produces a defective product is 1/3 . <br> What is the probability that the 1st defect is found during the first 5 inspections?

```python
a,b = 1,3 # 1/3 defective 
ins = 5
prob_defect = a/b  # .333 or 
prob_good = 1-prob_defect
prob = 0
for i in range(ins):
    prob+=(prob_good**i)*prob_defect
print('{:.3f}'.format(prob))
```
